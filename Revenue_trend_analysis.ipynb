```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Convert to datetime
df['MonthYear_dt'] = pd.to_datetime(df['MonthYear'], format='%b %Y')

# Convert to numeric (e.g., number of months since a reference date)
df['MonthYear_num'] = (df['MonthYear_dt'].dt.year - 2000) * 12 + df['MonthYear_dt'].dt.month

X = df[['MonthYear_num']]
y = df['Revenue']

model = LinearRegression()
model.fit(X, y)
```

```python
# Create future dataframe using the same feature name used in training
future = pd.DataFrame({'MonthYear_num': [13, 14, 15]})

# Predict
predictions = model.predict(future)
print("Next months revenue:", predictions)
```
